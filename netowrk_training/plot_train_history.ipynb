```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
```

# load history record

```python
with open("./train_result/history.json") as file:
        history = json.load(file)
```

# plot history

```python
fig, axes = plt.subplots(2,1, figsize=(16,8), dpi = 100)
plt.subplots_adjust(wspace =0, hspace = 0.3)
epoch = list(range(len(history['train_los'])))

axes[1].set_ylabel('Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_xticks(epoch)
axes[1].plot(epoch, history['train_los'], label='train', color = 'b')
axes[1].plot(epoch, history['test_los'], label='test' , color = 'r')
axes[1].legend()

axes[0].set_ylabel('HitRate')
axes[0].set_xlabel('Epoch')
axes[0].set_xticks(epoch)
axes[0].plot(epoch, history['train_acc'], label='train', color = 'b')
axes[0].plot(epoch, history['test_acc'], label='test' , color = 'r')
axes[0].legend()
```
